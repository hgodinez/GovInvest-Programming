File: MortalityDisability.csv

This file includes the mortality and disability rates of the CalPers demographics.
The variables included in the file are:


AA = Attained Age

PRNIDM = Pre-Retirement Mortality Non Industrial Death Male Assumptions

PRIDM =  Pre-Retirement Mortality Industrial Death Male Assumptions

PRNIDF = Pre-Retirement Mortality Non Industrial Death Female Assumptions

PRIDF =  Pre-Retirement Mortality Industrial Death Female Assumptions


PoRHM = Post-Retirement Mortality Healthy Recipients Male Assumptions

PoRNIDiM = Post-Retirement Mortality Non Industrially Disabled Male Assumptions

PoRIDiM =  Post-Retirement Mortality Industrially Disabled Male Assumptions

PoRHF = Post-Retirement Mortality Healthy Recipients Female Assumptions

PoRNIDiF = Post-Retirement Mortality Non Industrially Disabled Female Assumptions

PoRIDiF =  Post-Retirement Mortality Industrially Disabled Female Assumptions



DNIDM = Disability Non Industrial Disability Male Assumptions

DIDM =  Disability Industrial Disability Male Assumptions

DNIDF = Disability Non Industrial Disability Female Assumptions

DIDF =  Disability Industrial Disability Female Assumptions

														
														
Notes:														
    1) Pre-Retirement and Post-Retirement mortality rates include 20 years of projected on-going mortality improvement using Scale BB published by the Society of Actuaries.												
    2) Miscellaneous Plans usually have Industrial Death rates set to zero unless the agency has specifically contracted for Industrial Death benefits.  If so, each Non-Industrial Death rate shown above will be split into two components: 99% will become the Non-Industrial Death rate and 1% will become the Industrial Death rate.													
    3) The Miscellaneous Non-Industrial Disability rates are used for Local Prosecutors												4) Normally, Industrial Disability rates are zero for miscellaneous plans unless the agency has specifically contracted for Industrial Disability benefits.  If so, each miscellaneous non-industrial disability rate will be split into two components: 50% will become the Non-Industrial Disability rate and 50% will become the Industrial Disability rate.
